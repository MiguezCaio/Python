2) Crie uma função que receba uma string e indique se começa ou não com uma vogal.

```
def vogal(string):
  vogais=["a","e","i","o","u"]
  for i in vogais:
    if string[-len(string)]==i:
      return True
  return False
```

3) Crie uma função que receba strings ou objetos numéricos e indique, caso string, se a primeira letra é uma vogal ou não e, caso o input seja um número, indique se o número é positivo ou negativo.

```
def vogalnumero(info):
  if type(info)==str:
    vogais=["a","e","i","o","u"]
    for i in vogais:
      if info[-len(info)]==i:
        print(f'A informação dada, {info}, é uma string e começa em vogal')
    print(f"A informação dada, {info}, é uma string e não começa em vogal")
  elif type(info)== float or type(info)== int or type(info)== complex:
    if info>0:
      print(f"A informação dada, {info}, é um objeto numérico e é positivo")
    elif info==0:
      print(f"A informação dada, {info}, é um objeto numérico e é zero")
    else:
      print(f"A informação dada, {info}, é um objeto numérico e é negativo")
    return False
```

4) Crie uma função que receba um número e retorne seu fatorial.

```
def fatorial(numero):
  x=numero
  fatorial=1
  while x>1:
    fatorial=fatorial*x
    x=x-1
  return fatorial

```
